```python
grupo_tipo = datos.groupby('Tipo')['Valor'] 
```

```python
type(grupo_tipo)
```

```python
Q1 = grupo_tipo.quantile(.25)
Q3 = grupo_tipo.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ  
```

```python
datos_new2 = pd.DataFrame()
for tipo in grupo_tipo.groups.keys():
  eh_tipo = datos['Tipo'] == tipo 
  eh_dentro_limite = ((datos['Valor'] >= limite_inferior[tipo]) & (datos['Valor'] <= limite_superior[tipo]))
  seleccion2 = eh_tipo & eh_dentro_limite
  datos_seleccion = datos[seleccion2]
  datos_new2 = pd.concat([datos_new2, datos_seleccion])
```

```python
datos_new2.boxplot(['Valor'], by = ['Tipo']) 
```
